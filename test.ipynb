```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df1 = pd.read_csv('code/results/Cifar100_20clients_100Round_DirOriginal/accuracy/FedAvg_Random_8.0_309.csv')
df2 = pd.read_csv('code/results/Cifar100_20clients_100Round_Dir1/accuracy/FedAvg_Random_8.0_309.csv')

# Plot the data x axis within 100 epochs
plt.plot(np.arange(101), df1['FedAvg_Random_8.0_309'], label='Dirichlet Alpha=0.1')
plt.plot(np.arange(101), df2['FedAvg_Random_8.0_309'], label='Dirichlet Alpha=1.0')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Cifar100 20 clients 100 Rounds 40% Poisoned Clients (FedAvg)')
plt.legend()
plt.show()
```
